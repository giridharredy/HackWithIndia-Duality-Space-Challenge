```python
def move_yolo_files(source_dir, dest_dir):
    os.makedirs(os.path.join(dest_dir, 'images'), exist_ok=True)
    os.makedirs(os.path.join(dest_dir, 'labels'), exist_ok=True)
    for f in glob.glob(os.path.join(source_dir, 'images', '*')):
        shutil.move(f, os.path.join(dest_dir, 'images'))
    for f in glob.glob(os.path.join(source_dir, 'labels', '*')):
        shutil.move(f, os.path.join(dest_dir, 'labels'))
    print(f"Organized files from {source_dir} into {dest_dir}")

print("Organizing files into YOLO format with correct paths...")

move_yolo_files(os.path.join(EXTRACTED_TRAIN_PATH, 'train_1', 'train1'),
                os.path.join(YOLO_DATA_PATH, 'train'))
move_yolo_files(os.path.join(EXTRACTED_TRAIN_PATH, 'train_1', 'val1'),
                os.path.join(YOLO_DATA_PATH, 'val'))


move_yolo_files(os.path.join(EXTRACTED_TEST_PATH, 'Test'),
                os.path.join(YOLO_DATA_PATH, 'test'))

print("Dataset is ready for training")
```

```python

duality_yaml_content = f"""
path: {YOLO_DATA_PATH}
train: train/images
val: val/images
test: test/images

names:
  0: OxygenTank
  1: NitrogenTank
  2: FirstAidBox
  3: FireAlarm
  4: SafetySwitchPanel
  5: EmergencyPhone
  6: FireExtinguisher
"""
yaml_file_path = os.path.join(BASE_DIR, "data", "duality.yaml")
with open(yaml_file_path, "w") as f:
    f.write(duality_yaml_content)

print(f"âœ… duality.yaml created successfully at: {yaml_file_path}")
```

```python

model = YOLO("yolov8n.pt")

model.train(
    data=yaml_file_path,
    epochs=60,
    imgsz=640,
    batch=16,
    patience=15,
    project=f"{BASE_DIR}/models",
    name="duality_run_final",
    device=0 if torch.cuda.is_available() else 'cpu'
)
```

```python
import shutil
import os
import glob

print("--- Saving Results from Colab to Google Drive ---")

# --- Define Source and Destination Paths ---
# This is the folder in your Colab session where results are stored
COLAB_RESULTS_SOURCE_DIR = "/content/duality/models"

# This is the folder in your Google Drive where results will be saved
DRIVE_DESTINATION_DIR = "/content/drive/MyDrive/duality/final_results_and_models"

# Create the destination folder in your Drive if it doesn't exist
os.makedirs(DRIVE_DESTINATION_DIR, exist_ok=True)


# --- Copy the Latest Training Run Folder ---
# This folder contains your graphs and prediction images
try:
    latest_run_folder = sorted(glob.glob(os.path.join(COLAB_RESULTS_SOURCE_DIR, 'duality_run*')))[-1]
    run_name = os.path.basename(latest_run_folder)
    destination_path = os.path.join(DRIVE_DESTINATION_DIR, run_name)

    # Remove any old backup of the same run to avoid errors
    if os.path.exists(destination_path):
        shutil.rmtree(destination_path)

    print(f"Copying results folder '{run_name}' to your Google Drive...")
    shutil.copytree(latest_run_folder, destination_path)
    print(f"-> Successfully saved results to: {destination_path}")
except IndexError:
    print("WARNING: No training run folder found to back up.")


# --- Copy the Final Model Weights (.pt and .onnx files) ---
print("\nCopying final model weights...")
model_files = glob.glob(os.path.join(COLAB_RESULTS_SOURCE_DIR, '*.pt')) + glob.glob(os.path.join(COLAB_RESULTS_SOURCE_DIR, '*.onnx'))

if model_files:
    for model_file in model_files:
        shutil.copy(model_file, DRIVE_DESTINATION_DIR)
        print(f"-> Saved '{os.path.basename(model_file)}' to your Google Drive.")
else:
    print("WARNING: No .pt or .onnx model files found to back up.")

print("\nâœ… All results have been successfully saved to your Google Drive.")
```
